# Basics of Logistic Regression

```python
#import required libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
# Step 1 - Load Data
data_set = pd.read_csv("iphone_purchase_records.csv")
```

```python
sns.countplot(data=data_set, x='Purchase Iphone')
```

```python
y = data_set.pop("Purchase Iphone")
data_set
```

```python
# Step 2 - Convert Gender to number
from sklearn.preprocessing import LabelEncoder
labelEncoder_gender =  LabelEncoder()
data_set["Gender"] = labelEncoder_gender.fit_transform(data_set["Gender"])
```

```python
data_set
```

```python
# Step 3 - Split data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_set, y, test_size=0.25, random_state=0)
```

```python
# Step 4 - Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr= LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
prediction = lr.predict(X_test)
from sklearn import metrics
#cm = metrics.confusion_matrix(y_test, prediction) 
#print(cm)
accuracy = metrics.accuracy_score(y_test, prediction) 
print("Accuracy score:",accuracy)
```
